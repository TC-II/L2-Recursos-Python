```python
#Primero vamos a definir simbolos

A, B = sp.symbols('A, B_1') #Se puede poner formato latex
B
```

```python
#Podemos hacer operaciones con estos simbolos

C = A + 3

B = A*B     # Tener en cuenta que esto pisa la variable!!!!
B
# La referencia a B esta guardada en una expresion pero no la podemos acceder porque se reemplazo la referencia.
```

```python
#Podemos reemplazar valores dentro de las expresiones
C.subs(A, 4) #Esto no lo reemplaza hasta que no lo igualamos a una expresion
C1 = C.subs(A, 4)

display(C1)
display(C)
```

```python
#Simplificaciones donde funcionan
x = sp.symbols('x')
num =  x**2 + 9 + 1/x + x**5 + x**2
den = (x -1)**2 + 4 + x + 1
fun = num/den
display(fun,"\n"), print(" ")

fun_simply = sp.simplify(fun)
display(fun_simply), print(" ")

fun_cancel = sp.cancel(fun,x)
display(fun_cancel), print(" ")

fun_expand = sp.expand(fun,x)
display(fun_expand)
```

```python
#Ejemplo: No inversor con Avol
Vo, Vi, Ao, wd, w, R1, R2= sp.symbols('V_o, V_i, A_0, w_d, w, R_1, R_2', real=True)
# Es una buena práctica poner reales y complejos por separado. Tambien se puede definir "positive = True".
s = sp.symbols('s', complex=True)
```

```python
Avol = Ao/(1+s/wd)
kni = 1 + R1/R2
Vo =  Vi * kni *  1/(1 + kni/Avol)
display(Vo)
Vo_s = sp.simplify(Vo)      #No siempre es bueno simplificar. Hay casos patológicos en los que es peor simplificar.
display(Vo_s)
```
